```python
from sklearn.datasets import make_blobs
```

```python
X,notuse=make_blobs(n_samples=500)  #it generates input and output data randomly
```

```python
X
```

```python
notuse
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1]) #ploting the points from 1st and second columns
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)  #default n_clusters was 8-->shift+tab
```

```python
km.fit(X)
```

```python
km.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1])
plt.scatter(9.67082804, -9.62923393)
plt.scatter(5.37315862,  4.43501948)
plt.scatter(7.04988307,  7.68050438)
```

```python
km.labels_  #since there are 3 lusters labels are 0,1,2
```
