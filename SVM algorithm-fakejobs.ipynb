```python
#title column contains a lot of random values that is not useful hence we will try to cvert it into 
#categories by custom phyton code.
```

```python
IT_title=['developer','data','engineer','software','web','analyst','network','designer','frontend'
          ,'backend','IT','php','technical']
admin_title=['manager','management','account','sales','marketing','product','executive','administrator'
             ,'supervisor','program']
```

```python
for i in range(df.shape[0]):
    current_title=df['title'].values[i]
    list_title=current_title.split(' ')
    simplified_title=''
    
    for each in list_title:
        if each.lower() in IT_title:
            simplified_title='IT'
            break
        if each.lower() in admin_title:  #.lower() is used so that titles with lower case that we have assigned in the above cell can be compared to the original ones by just converting it to lower case
                                         #else that they are ignored-Like developer/Developer.
            simplified_title='Management'
            break;
        else:
            simplified_title='Others'
    df['title'].values[i]=simplified_title       
```

```python
#let us balance the dataset and then chek for the accuracy again
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus=RandomUnderSampler()
```

```python
X=df[['title', 'location', 'salary_range', 'telecommuting',
       'has_company_logo', 'has_questions', 'employment_type',
       'required_experience', 'required_education', 'industry', 'function']]
Y=df[['fraudulent']]  #.values is not used since we want it as a dataframe for balancing in rus.
```

```python
X_bal,Y_bal=rus.fit_resample(X,Y)
```

```python
Y_bal['fraudulent'].value_counts() #balanced
```

```python
X_bal=X_bal.values  #concerting to numpy array
Y_bal=Y_bal.values
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_bal,Y_bal,train_size=0.75)
```

```python
svm.fit(X_train,Y_train)
rfc.fit(X_train,Y_train)
```

```python
Y_pred_rfc2=rfc.predict(X_test)
Y_pred_svm2=svm.predict(X_test)
```

```python
accuracy_score(Y_test,Y_pred_svm2)
```

```python
accuracy_score(Y_test,Y_pred_rfc2)  #balancing usually reduces the accuracy.
                                   #but this is prefered since it is not biased.
```
