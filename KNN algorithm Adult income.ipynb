```python
df=df[['age', 'workclass', 'education.num', 'marital.status', 'occupation',
       'relationship', 'race', 'sex', 'hours.per.week',
       'income']]
```

```python
X=df[['age', 'workclass', 'education.num', 'marital.status', 'occupation',
       'relationship', 'race', 'hours.per.week',
       'income']].values
Y=df[['sex']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.75)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=6)
```

```python
knn.fit(X_train,Y_train.ravel())   #.ravel is used to avoid a warning that comes since all values of Y_train is stored in seperate arrays and not 1 array for all the values.
```

```python
Y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_pred)  #neighbors value with 6 gives the max accuracy value , find by trial and error, try changing neighbor values and match for the accuracy rates.
```

```python
from imblearn.under_sampling import RandomUnderSampler   #balancing the dataset.
```
