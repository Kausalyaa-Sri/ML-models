```python
X=df[['fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel',
       'enginelocation', 'wheelbase', 'carlength', 'carwidth', 'carheight',
       'curbweight', 'enginetype', 'cylindernumber', 'enginesize',
       'fuelsystem', 'boreratio', 'stroke', 'compressionratio', 'horsepower',
       'peakrpm', 'citympg', 'highwaympg']].values
Y=df[['price']].values    #.values is to convert the dataframe to numpy array so that my algorithm accepts my values.
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)#fit function is used to train the model
```

```python
Y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_absolute_error(Y_test,Y_pred)
```

```python
mean_squared_error(Y_test,Y_pred)
```
